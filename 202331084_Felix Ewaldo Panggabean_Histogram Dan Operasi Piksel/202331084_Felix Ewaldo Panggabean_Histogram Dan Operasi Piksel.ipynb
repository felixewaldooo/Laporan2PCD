```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('tahusumedang.png.jpg')

if img is None:
    print("Citra tidak dapat dimuat. Pastikan path gambar benar.")
else:
    print("Citra berhasil dimuat.")
```

```python
# Felix Ewaldo Panggabean 202331084 Pengolahan Citra Digital (C)
img.shape
```

```python
[baris, kolom] = img.shape[:2]
```

```python
# Felix Ewaldo Panggabean 202331084
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# Cara 1: Menggunakan imshow dan hist pada subplot
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].imshow(rgb) # Menampilkan gambar pada subplot [0, 0]
axs[0, 0].set_title("Citra RGB") # Menambahkan judul untuk subplot [0, 0]
gray = cv2.cvtColor(rgb, cv2.COLOR_RGB2GRAY)
axs[0, 1].hist(rgb.ravel(), 256, [0, 256], color='pink') # Histogram grayscale pada subplot [0, 1]
axs[0, 1].set_title("Histogram Grayscale") # Menambahkan judul untuk subplot [0, 1]

# Cara 2: Menggunakan cv2.calcHist dan plot pada subplot
hist = cv2.calcHist([rgb], [0], None, [256], [0, 256]) # Menghitung histogram channel merah
axs[1, 0].imshow(rgb) # Menampilkan gambar pada subplot [1, 0]
axs[1, 0].set_title("Citra RGB (Ulang)") # Menambahkan judul untuk subplot [1, 0]

axs[1, 1].plot(hist, color='blue') # Plot histogram channel merah pada subplot [1, 1]
axs[1, 1].set_title("Histogram Channel Merah") # Menambahkan judul untuk subplot [1, 1]

plt.tight_layout() # Untuk mengatur layout agar tidak tumpang tindih
plt.show()
# Felix Ewaldo Panggabean_202331084
```

```python
beta = 40
# Felix Ewaldo Panggabean 202331084
# Membuat array untuk menyimpan hasil peningkatan kecerahan
baris, kolom, _ = rgb.shape
citra_cerah = np.zeros((baris, kolom, 3))

# Menghitung peningkatan kecerahan untuk setiap piksel
for y in range(baris):
    for x in range(kolom):
        # Hitung kecerahan baru untuk setiap channel (R, G, B)
        gyx = rgb[y, x] + beta
        # Batasi nilai agar tetap dalam rentang [0, 255]
        gyx = np.clip(gyx, 0, 375)
        citra_cerah[y, x] = gyx

# Ubah tipe data menjadi uint8
citra_cerah = citra_cerah.astype(np.uint8)

# Menampilkan hasil menggunakan subplot
fig, axs = plt.subplots(2, 2, figsize=(15, 5))

# Subplot 1: Gambar asli
axs[0, 0].imshow(rgb)
axs[0, 0].set_title("Citra RGB")
axs[0, 0].axis('off') # Matikan sumbu koordinat

# Subplot 2: Histogram citra asli
axs[0, 1].hist(rgb.ravel(), 256, [0, 256], color='pink')
axs[0, 1].set_title("Histogram RGB")

# Subplot 3: Gambar dengan peningkatan kecerahan
axs[1, 0].imshow(citra_cerah)
axs[1, 0].set_title("Citra dengan Peningkatan Kecerahan")
axs[1, 0].axis('off') # Matikan sumbu koordinat
```

```python
# Faktor kontras
alpha = 1.2

# Felix Ewaldo Panggabean 202331084
# Membuat array untuk menyimpan hasil kontras
citra_Kontras = np.zeros((baris, kolom, 3))

# Menghitung kontras untuk setiap piksel
for i in range(baris):
    for j in range(kolom):
        # Hitung kontras untuk setiap channel (R, G, B)
        gyx = rgb[i, j] * alpha
        # Batasi nilai agar tetap dalam rentang [0, 255]
        gyx = np.clip(gyx, 0, 255)
        citra_Kontras[i, j] = gyx

# Ubah tipe data menjadi uint8
citra_Kontras = citra_Kontras.astype(np.uint8)

# Menampilkan hasil menggunakan subplot
fig, axs = plt.subplots(2, 2, figsize=(15,5))

# Subplot 1: Gambar asli
axs[0, 0].imshow(rgb)
axs[0, 0].set_title("Citra Asli")
axs[0, 0].axis('off') # Matikan sumbu koordinat

# Subplot 2: Histogram citra asli
axs[0, 1].hist(rgb.ravel(), 256, [0, 256], color='gray')
axs[0, 1].set_title("Histogram Citra Asli")

# Subplot 3: Gambar dengan kontras ditingkatkan
axs[1, 0].imshow(citra_Kontras)
axs[1, 0].set_title("Citra dengan Kontras")
axs[1, 0].axis('off') # Matikan sumbu koordinat
```
